```python
import numpy as np
import scipy.linalg as LA
from itertools import permutations

def tuple_to_matrix(tup):
    return np.array(tup).reshape(2, 2)

perms = list(permutations([0, 1, 0, 1]))

unique_matrices = {}
all_methods = []

for perm in perms:
    matrix = tuple_to_matrix(perm)
    det = LA.det(matrix)  
    all_methods.append((matrix, det))
    matrix_str = np.array_str(matrix)  
    unique_matrices[matrix_str] = det

print("All 24 filling methods and their determinants:")
for method, det in all_methods:
    print(method)
    print("Determinant:", det)
    print("---")

det_counts_all_methods = {}
for _, det in all_methods:
    det_counts_all_methods[det] = det_counts_all_methods.get(det, 0) + 1

print("\nUnique matrix combinations and their determinants:")
for matrix_str, det in unique_matrices.items():
    print(matrix_str)
    print("Determinant:", det)
    print("---")

det_counts_unique_matrices = {}
for det in unique_matrices.values():
    det_counts_unique_matrices[det] = det_counts_unique_matrices.get(det, 0) + 1

print("\nCount of each unique determinant for all 24 methods:")
for det, count in det_counts_all_methods.items():
    print(f"Determinant: {det}, Count: {count}")

print("\nCount of each unique determinant for the unique matrices:")
for det, count in det_counts_unique_matrices.items():
    print(f"Determinant: {det}, Count: {count}")
```

```python
import random
def random_move(grid):
    
    available_positions = [(i, j) for i in range(grid.shape[0]) for j in range(grid.shape[1]) if grid[i, j] == -1]
    return random.choice(available_positions) if available_positions else None

def initialize_grid(n):
    """Initialize an n x n grid with -1, indicating empty spaces."""
    return np.full((n, n), -1)

def find_strategic_move(grid, player, strategy):
    """Find a strategic move based on the given strategy."""
    available_moves = np.argwhere(grid == -1)
    if strategy == "follow":
       
        player_moves = np.argwhere(grid == player)
        for move in available_moves:
            if any((move[0] == pm[0] or move[1] == pm[1]) for pm in player_moves):
                return move
    elif strategy == "avoid":
        
        opponent_moves = np.argwhere(grid == (1 - player))
        for move in available_moves:
            if all(move[0] != om[0] and move[1] != om[1] for om in opponent_moves):
                return move
    
    return random_move(grid)


def play_game(n=2, my_strategy="avoid", friend_strategy="random", I_start=True):
    grid = initialize_grid(n)
    player = 0 if I_start else 1  
    
    while -1 in grid.flatten():  
        current_strategy = my_strategy if player == 0 else friend_strategy
        move = find_strategic_move(grid, player, current_strategy) if current_strategy != "random" else random_move(grid)
        if move is not None:
            grid[move[0], move[1]] = player
        player = 1 - player  
    
    determinant = np.linalg.det(grid.astype(np.float64))
    win = determinant == 0
    return win, determinant

win, determinant = play_game()
win, determinant

```

```python
def simulate_games(n=2, I_start=True, simulations=100):
    """Simulate the game for 100 times for both scenarios: I start or my friend starts."""
    wins_count = 0 
    
    for _ in range(simulations):
        win, determinant = play_game(n=n, I_start=I_start)
        if win:
            wins_count += 1
            
    return wins_count


wins_count_me_start = simulate_games(I_start=True)


wins_count_friend_start = simulate_games(I_start=False)

wins_count_me_start, wins_count_friend_start
```

```python
import numpy as np
from itertools import combinations

def generate_matrices_with_det_zero(n):
    total_elements = n * n
    winning_matrices = []
    
    for num_ones in [4, 5]:
        positions = range(total_elements)
        for ones_positions in combinations(positions, num_ones):
            matrix = np.zeros(total_elements, dtype=int)
            matrix[list(ones_positions)] = 1
            reshaped_matrix = matrix.reshape(n, n)
            
           
            if np.isclose(np.linalg.det(reshaped_matrix), 0, atol=1e-10):
                winning_matrices.append(reshaped_matrix)
                
    return winning_matrices

n = 3
winning_matrices = generate_matrices_with_det_zero(n)
```

```python
def simulate_games_adjusted(n, iterations, I_start):
    wins = 0
    for _ in range(iterations):
        
        my_strategy = "follow" if I_start else "avoid"
        friend_strategy = "random"
        
       
        grid = initialize_grid(n)
        player = 0 if I_start else 1 
        while -1 in grid.flatten():
            current_strategy = my_strategy if player == 0 else friend_strategy
            move = find_strategic_move(grid, player, current_strategy) if current_strategy != "random" else random_move(grid)
            if move is not None:
                grid[move[0], move[1]] = player
            player = 1 - player  
        
        determinant = np.linalg.det(grid.astype(np.float64))
        if np.isclose(determinant, 0, atol=1e-10):
            wins += 1
    
    return wins / iterations
```

```python

n = 3  
iterations = 1000 

win_rate_I_start_3x3 = simulate_games_adjusted(n, iterations, True) 
win_rate_friend_starts_3x3 = simulate_games_adjusted(n, iterations, False)  

win_rate_I_start_3x3, win_rate_friend_starts_3x3
```

```python
I_start = []
friend_start = []
for i in range(2, 10):
    I_start.append(simulate_games_adjusted(i, 1000, True))
    friend_start.append(simulate_games_adjusted(i, 1000, False))
    
print(I_start)
print(friend_start)
```
